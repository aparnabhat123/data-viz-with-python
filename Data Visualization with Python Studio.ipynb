```python
# Here is the setup of the dataframe. Feel free to use .info() or .head() to get a better understanding of what is inside the dataframe!

import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
from matplotlib import style

df = pd.read_csv("books.csv", error_bad_lines=False)

years = []

for y in df['publication_date']:
    date = y.split("/")
    years.append(date[2])

df["year"] = years
```

With your dataframe set up, time to start creating visualizations!

```python
# Use your Python skills to create a visualization here.
df['year'].value_counts()
```

```python
# Use your Python skills to create a visualization here.
import matplotlib.pyplot as plt
#style.use('ggplot')
plt.rcParams['figure.figsize'] = (30,20) 
df['year'].value_counts()[:20].plot(kind='bar' , color = 'blue' , title = 'Number of books published each year' , xlabel = 'year' , ylabel = 'Total Number of books')
```

```python
# Use your Python skills to create a visualization here.
import matplotlib.pyplot as plt
style.use('ggplot')
plt.rcParams['figure.figsize'] = (30,10) 

plt.xlabel('Years',color = 'blue')
plt.ylabel('Number of books',color = 'blue')
plt.xticks(color = 'black')
plt.yticks(color = 'black')
plt.title('Number of books published each year',color = 'blue')
df['year'].value_counts().plot(kind='bar' , color = 'blue')
```

```python
#df['ratings_count'] = df['ratings_count']/len(df['ratings_count'])
#df['text_reviews_count'] = df['text_reviews_count']/len(df['text_reviews_count'])

df_group = df.tail(20).plot.bar(x = 'year', y = ['ratings_count', 'text_reviews_count'])
#df_group.plot(edgecolor = 'white')
plt.title('Total Immigrants to Canada from 1980-2013',color = 'black')
plt.xticks(color = 'black')
plt.yticks(color = 'black')
plt.xlabel('Year',color = 'black')
plt.ylabel('Number of Immigrants',color = 'black')
plt.legend(title = 'Country', fontsize = 20)
plt.savefig('bar_grouped.png')

plt.show()

# This infers as ratings_count increases , the text_review_count also increases 
```
